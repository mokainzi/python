```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
df_filepath = 'bestsellers with categories.csv'
df = pd.read_csv(df_filepath)
```

```python
sns.distplot(df['Price']) #, kde=False)
```

```python
sns.distplot(df['User Rating']) #, kde=False)
```

```python
sns.distplot(df['Reviews']) #, kde=False)
```

```python
sns.barplot(x=df['Genre'], y=df['Price'])
```

```python
plt.figure(figsize=(10,7))
sns.lmplot(data=df,
           x='User Rating', 
           y='Price', 
#            fit_reg=False,
           hue='Genre')
plt.title('User Rating By Book Price')
```

```python
plt.figure(figsize=(10,7))
sns.lmplot(data=df,
           x='Reviews', 
           y='Price', 
#            fit_reg=False,
           hue='Genre')
plt.title('Reviews By Book Price')
plt.xticks(rotation=45)
```

```python
plt.figure(figsize=(10,7))
sns.swarmplot(x=df['Genre'], y=df['Price'])
plt.title('Book Genre By Price')
```

```python
plt.figure(figsize=(10,7))
sns.swarmplot(x=df['Genre'], y=df['Reviews'])
plt.title('Book Genre By Reviews')
```

```python
plt.figure(figsize=(15,7))

sns.lineplot(x='Year', y='Price', data=df, label='Price')
sns.lineplot(x='Year', y='User Rating', data=df, label='User Rating')
```
