```python
import numpy as np
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import NumeralTickFormatter

output_notebook()
```

```python
df_filepath = 'bestsellers with categories.csv'
df = pd.read_csv(df_filepath)
```

```python
df
```

```python
year_avg = df.groupby('Year').mean()
```

```python
year_avg
```

```python
# Here is a list of categorical values (or factors)
genre = df.groupby('Genre')['Genre'].count().index.array                    # ISSUE HERE 
                                                                            # using unique() shows unmatched results
counts = df.groupby('Genre')['Genre'].count()

# Set the x_range to the list of categories above
p = figure(x_range=genre, plot_height=300, title="Genre Counts")

# Categorical values can also be used as coordinates
p.vbar(x=genre, top=counts, width=0.9)

# Set some properties to make the plot look better
p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
```

The data contains more Non fiction than Fiction books.

```python
x = year_avg.index
y = year_avg['Price']

p = figure(title="Book Prices Over Time", plot_height=350, plot_width=800)

p.xgrid.grid_line_color=None
p.ygrid.grid_line_alpha=0.5

p.xaxis.axis_label = x.name
p.yaxis.axis_label = y.name

p.yaxis.formatter = NumeralTickFormatter(format="$0")

p.line(x, y, line_color='firebrick')

show(p)
```

Looks like Amazon's top 50 best selling books' prices were decreasing. I wonder if there's a relation between prices and reviews, or user ratings..

```python
x = year_avg['Reviews']
y = year_avg['Price']


par = np.polyfit(x, y, 1, full=True)
slope=par[0][0]
intercept=par[0][1]
y_predicted = [slope*i + intercept  for i in x]


p = figure(title='Reviews vs. Book Price', plot_width=400, plot_height=400)

p.xaxis.axis_label = x.name
p.yaxis.axis_label = y.name

p.yaxis.formatter = NumeralTickFormatter(format="$0")

p.circle(x, y, size=10, line_color=None)
p.line(x,y_predicted,color='orange')

show(p)
```

The data shows that on average, the lower price the books are the higher reviews they get.

```python
x = year_avg['User Rating']
y = year_avg['Price']


par = np.polyfit(x, y, 1, full=True)
slope=par[0][0]
intercept=par[0][1]
y_predicted = [slope*i + intercept  for i in x]


p = figure(title='User Rating vs. Book Price', plot_width=400, plot_height=400)

p.xaxis.axis_label = x.name
p.yaxis.axis_label = y.name

p.yaxis.formatter = NumeralTickFormatter(format="$0")

p.circle(x, y, size=10, line_color=None, fill_color="firebrick")
p.line(x,y_predicted,color='orange')

show(p)
```

The data shows that on average, the lower price the books are the higher rating they get.
